```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Year': np.arange(2010, 2021),
    'Sales': np.random.randint(200, 1000, 11),
    'Profit': np.random.randint(50, 400, 11),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 11)
})

# Display the data
print(data)

# --- LINE PLOT ---
plt.figure(figsize=(8,5))
plt.plot(data['Year'], data['Sales'], marker='o', color='b', label='Sales')
plt.plot(data['Year'], data['Profit'], marker='s', color='g', label='Profit')
plt.title('Yearly Sales and Profit')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# --- SCATTER PLOT ---
plt.figure(figsize=(8,5))
plt.scatter(data['Sales'], data['Profit'], color='purple')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()

# --- HISTOGRAM ---
plt.figure(figsize=(8,5))
sns.histplot(data['Sales'], bins=10, kde=True, color='orange')
plt.title('Distribution of Sales')
plt.show()

# --- BOX PLOT ---
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Profit', data=data, palette='Set2')
plt.title('Profit Distribution by Region')
plt.show()
```
